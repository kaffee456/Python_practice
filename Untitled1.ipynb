```python
import numpy as np
from matplotlib import pyplot as plt
from mpl_toolkits.mplot3d import Axes3D

r = 3 # 
theta_1_0 = np.linspace(0, np.pi, 100) # _1[0,/2]
theta_2_0 = np.linspace(0, np.pi*2, 100) # _2[0,/2]
theta_1, theta_2 = np.meshgrid(theta_1_0, theta_2_0) # 
x = np.cos(theta_2)*np.sin(theta_1) * r # x
y = np.sin(theta_2)*np.sin(theta_1) * r # y
z = np.cos(theta_1) * r # z

fig = plt.figure() # 
ax = fig.add_subplot(111, projection="3d") # 3D
ax.plot_surface(x,y,z) # 
plt.xlim([0,10])
plt.ylim([10,0]) # Y
ax.set_zlim([0,10])
plt.xlabel("X")
plt.ylabel("Y")
ax.set_zlabel("Z")
plt.show()
```
